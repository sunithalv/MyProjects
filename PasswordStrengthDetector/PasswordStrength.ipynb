```python
#Function to seperate char in password
def word_divide_char(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#Since we do not have sentence here we need to work on char by char so we create a function for the same
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
X=vectorizer.fit_transform(X)
```

```python
X.shape
```

```python
vectorizer.vocabulary_
```

```python
X
```

```python
data.iloc[0,0]
```

```python
feature_names = vectorizer.get_feature_names()
 
#get tfidf vector for first document
first_document_vector=X[0]
 
#print the scores
df = pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```

### Model Building

```python
## Logistics Regression
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)  #splitting
```

```python
classifier=[]
classifier.append(LogisticRegression(penalty='l2',multi_class='ovr'))
classifier.append(LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg',n_jobs=-1))
classifier.append(xgb.XGBClassifier(n_jobs=-1))
classifier.append(MultinomialNB())
```

```python
result=[]
for model in classifier:
    a=model.fit(X_train,y_train)
    result.append(a.score(X_test,y_test))
```

```python
result_df=pd.DataFrame({'score':result,'algorithms':['LogisticReg ovr','LogisticReg multinomial','XGBoost','NaiveBayes']})
```

```python
result_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,8))
a=sns.barplot(data=result_df,y='score',x='algorithms')
a.set_label('accuracy')
```

### XGBoost

```python
xgb_classifier=xgb.XGBClassifier()
```

```python
xgb_classifier.fit(X_train, y_train) #training
```

```python
pred=xgb_classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```
